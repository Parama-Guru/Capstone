```python
from flask import Flask, render_template
from firebase_admin import db

app = Flask(__name__)

@app.route('/')
def index():
    # Example: Reading data from Firebase Realtime Database
    ref = db.reference('/')
    data = ref.get()

    return render_template('index.html', data=data)

if __name__ == '__main__':
    app.run(debug=True)
```
